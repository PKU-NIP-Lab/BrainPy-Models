# Adaptive Exponential Integrate-and-Fire model

Adaptive Exponential Integrate-and-Fire neuron model is a spiking model, describes single neuron behavior and can generate many kinds of firing patterns by tuning parameters.

```python
import brainpy as bp
import numpy as np
import brainmodels
import matplotlib.pyplot as plt

bp.backend.set(backend='numpy', dt=.005)
```

### Tonic

```python
tonic = dict(tau_m = 20, a = 0., tau_w = 30, b = 60, u_r = -55)
adapting = dict(tau_m = 20, a = 0., tau_w = 100, b = 5., u_r = -55)
init_bursting = dict(tau_m = 5., a = .5, tau_w = 100, b = 7., u_r = -51)
bursting = dict(tau_m = 5, a = -0.5, tau_w = 100, b = 7., u_r = -46)
transient = dict(tau_m = 10, a = 1., tau_w = 100, b = 10., u_r = -60)
delayed = dict(tau_m = 5, a = -1., tau_w = 100, b = 10., u_r = -60)

pars = tonic

threshold = -50
I_ext = 65
duration = 600
group = brainmodels.neurons.AdExIF(size = 1, monitors = ['V'], 
                                    a=pars['a'], b=pars['b'], R=.5, 
                                    tau=pars['tau_m'], tau_w=pars['tau_w'],
                                    V_reset=pars['u_r'], V_rest=-70, V_th=-30, 
                                    V_T=threshold, delta_T=2.)
group.run(duration, inputs=('input', I_ext))
bp.visualize.line_plot(group.mon.ts, group.mon.V, 
                        ylim=(-65., -30.), title='tonic',
                        show=True)
```
